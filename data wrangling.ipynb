### `query( )` - Filtering Rows in your Data Set

Filter row with the year 1972 and with a life expectancy below average, and with the country either to be Bolivai OR Angola

```python
(
    gapminder
    .filter(items=['country','year','lifeExp'])
    .query('year==1972')
    .query("lifeExp < lifeExp.mean()")
    .query("country == 'Bolivia' | country == 'Angola'")
).head(rows)
```

### `assign( )` and `lambda` - Generate new Columns in your Data Set
Create a column that combines continent and country information, and another column that shows the rounded lifeExp information.

```python
(
    gapminder
    .sort_values(['year','pop'])
    .assign(
        conn_country = lambda x: x.continent + " - " + x.country,
        rn_lifeExp = lambda x: x.lifeExp.round(0).astype(int))
    .filter(items = ['continent','country','conn_country','lifeExp','rn_lifeExp'])
    
    ).head(rows)
```

### `agg( ) ` - Create Summary Calculations in your Data Set
For the whole data set calculate mean and standard deviation for population and life expectations.

```python
(
    gapminder
    .agg(
        pop_mean = ('pop', 'mean'),
        pop_std = ('pop', 'std'),
        le_mean = ('lifeExp','mean'),
        le_sd = ('lifeExp','std')
    )
)
```

### `groupby( )` - Group your DataSet and Create Summary Calculations

the `agg()` function is only so useful without the `groupby()` function. Using both together is a powerful way to create new data sets. In the example below, I will group the data set by Continent and then I will create summaries for population and lifeExp.

```python
(
    gapminder
    .agg(
        pop_mean = ('pop', 'mean'),
        pop_std = ('pop', 'std'),
        le_mean = ('lifeExp','mean'),
        le_sd = ('lifeExp','std')
    )
)
```

```python
(
    gapminder
    .query("continent == 'Africa'")
    .groupby(['continent','year'])
    .agg(
        pop_mean = ('pop', 'mean'),
        pop_std = ('pop', 'std'),
        le_mean = ('lifeExp','mean'),
        le_sd = ('lifeExp','std')
    )
)
```
